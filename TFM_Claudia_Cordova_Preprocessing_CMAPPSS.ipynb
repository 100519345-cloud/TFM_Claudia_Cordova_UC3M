```python
import pandas as pd

n_column = ['engine_id', 'cycle'] + \
               ['op_setting_1', 'op_setting_2', 'op_setting_3'] + \
               [f's{i}' for i in range(1, 22)]

df = pd.read_csv('train_FD001.txt', sep=' ', header=None)
df.dropna(axis=1, inplace=True)
df.columns = n_column

df.head()
```

```python
# To calculate the maximum cicle per motor
rul_df = df.groupby('engine_id')['cycle'].max().reset_index()
rul_df.columns = ['engine_id', 'max_cycle']

# To unify the actual cicle with the maximum
df = df.merge(rul_df, on='engine_id', how='left')

# To calculate the RUL as a difference
df['RUL'] = df['max_cycle'] - df['cycle']
df.drop('max_cycle', axis=1, inplace=True)

df[['engine_id', 'cycle', 'RUL']].head()
```

```python
# This part is to see the standard deviation of each sensor
sensor_cols = [f's{i}' for i in range(1, 22)]
df[sensor_cols].std().sort_values()
```

```python
# The same but for graphics
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = df[sensor_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix of Sensor Readings')
plt.show()
```

```python
# Select only the sensors that I analyze in the correlation matrix
selected_sensors = ['s2', 's3', 's4', 's7', 's8', 's9', 's11', 's12', 's13', 's14', 's15', 's17', 's20', 's21']
```

```python
# New Dataframe with select sensors
features = ['engine_id', 'cycle', 'RUL'] + selected_sensors
df_selected = df[features]
df_selected.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# It is for defining the columns to scale
columns_to_scale = selected_sensors
scaler = MinMaxScaler()

# Aplying the scale
df_selected[columns_to_scale] = scaler.fit_transform(df_selected[columns_to_scale])

# Verify the first value
df_selected.head()
```

```python
# How change the sensors respect the RUL
import matplotlib.pyplot as plt

sensor_to_plot = 's2'
plt.figure(figsize=(10, 6))
plt.plot(df_selected['RUL'], df_selected[sensor_to_plot], '.', alpha=0.5)
plt.xlabel('Remaining Useful Life (RUL)')
plt.ylabel(sensor_to_plot)
plt.title(f'{sensor_to_plot} vs RUL')
plt.grid(True)
plt.show()
```
